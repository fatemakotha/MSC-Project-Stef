#**Visualizations**

```python
#Visualize counts for 0 and 1 for dev dataset
sns.set_style("whitegrid")
plt.figure(figsize=(18, 6))


plt.subplot(1, 5, 1)
sns.countplot(x="label", data=dev_df)
plt.title("Dev Dataset")

plt.tight_layout()
plt.show()
```

```python
#Visualize counts for 0 and 1 for test dataset
sns.set_style("whitegrid")
plt.figure(figsize=(18, 6))

plt.subplot(1, 5, 2)
sns.countplot(x="label", data=test_df)
plt.title("Test Dataset")

plt.tight_layout()
plt.show()
```

```python
#Visualize counts for 0 and 1 for train few shot dataset
sns.set_style("whitegrid")
plt.figure(figsize=(18, 6))


plt.subplot(1, 5, 3)
sns.countplot(x="label", data=train_few_shot_df)
plt.title("Train Few Shot Dataset")

plt.tight_layout()
plt.show()
```

```python
#Visualize counts for 0 and 1 for train one shot dataset
sns.set_style("whitegrid")
plt.figure(figsize=(18, 6))


plt.subplot(1, 5, 4)
sns.countplot(x="label", data=train_one_shot_df)
plt.title("Train One Shot Dataset")

plt.tight_layout()
plt.show()
```

```python
#Visualize counts for 0 and 1 for train zero shot dataset
sns.set_style("whitegrid")
plt.figure(figsize=(18, 6))


plt.subplot(1, 5, 5)
sns.countplot(x="label", data=train_zero_shot_df)
plt.title("Train Zero Shot Dataset")

plt.tight_layout()
plt.show()
```
